```python
#Data PreProcessing : all entries except property State = India
df = df[df['State'] != "India"]

#describe the dataset
display(df.head())
```

```python
#Million , Billion Formatting
def human_format(num):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    return '{:.2f}{}'.format(num, ['', 'K', 'M', 'B', 'T'][magnitude])

```

```python

import pandas as pd

# Set the display format to avoid scientific notation
pd.options.display.float_format = '{:.0f}'.format

#b.Number of persons statewise vaccinated for first dose in india

first_dose_by_state = df.groupby('State')['First Dose Administered'].sum()
formatted_output = first_dose_by_state.apply(human_format)

print(formatted_output)
```

```python
#c. Number of persons Statewise vaccinated for second dose in india
second_dose_by_state  = df.groupby('State')['Second Dose Administered'].sum()
formatted_output_second = second_dose_by_state.apply(human_format)

print(formatted_output_second)
```

```python
#d.Number of Males vaccinated
males_vaccinated = df['Male (Doses Administered)'].sum()

formatted_output_males = human_format(males_vaccinated)

print("Total Males vaccinated : " , formatted_output_males)
```

```python
#e.Number of Females Vaccinated
females_vaccinated = df['Female (Doses Administered)'].sum()

formatted_output_females = human_format(females_vaccinated)

print("Total Females vaccinated : " , formatted_output_females)
```
